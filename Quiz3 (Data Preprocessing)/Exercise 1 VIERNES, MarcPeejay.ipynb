#### 4. Handling missing data
#### If column contains more than 70% null values, remove that column.  Otherwise, remove rows with null values.

```python
#remove Null values
# your code starts here
for column_name in df.columns:
    if (df[column_name].isna().sum() / nrows) > 0.7:
        df.drop(column_name, axis=1, inplace=True)
        
df = df.dropna()
# your code ends here

#Get total number of null in each column
# your code starts here
df.isna().sum()
# your code ends here
```

```python
# Update the LVM column by grouping the regions to Luzon, Visayas, Mindanao 
# You can use replace function from pandas
# your code starts here
# Reference: https://www.philatlas.com/regions.html
region_to_LVM_mapping = {
    'REGION I (ILOCOS REGION)': 'Luzon',
    'REGION II (CAGAYAN VALLEY)': 'Luzon',
    'REGION III (CENTRAL LUZON)':  'Luzon',
    'REGION IV-A (CALABAR ZON)': 'Luzon',
    'REGION IV-B (MIMAROPA)': 'Luzon',
    'REGION V (BICOL REGION)': 'Luzon',
    'CORDILLERA ADMINISTRA TIVE REGION (CAR)': 'Luzon',
    'NATIONAL CAPITAL REGION (NCR)': 'Luzon',
    'REGION VI (WESTERN VISAYAS)': 'Visayas',
    'REGION VII (CENTRAL VISAYAS)': 'Visayas',
    'REGION VIII (EASTERN VISAYAS)': 'Visayas',
    'REGION IX (ZAMBOANGA PENINSULA)': 'Mindanao',
    'REGION X (NORTHERN MINDANAO)': 'Mindanao',
    'REGION XI (DAVAO REGION)': 'Mindanao',
    'REGION XII (SOCCSKSA RGEN)': 'Mindanao',
    'REGION XIII (CARAGA)': 'Mindanao',
    'AUTONOMOUS REGION IN MUSLIM MINDANAO (ARMM)': 'Mindanao',
    'BARMM': 'Mindanao'
}

df['LVM'].replace(region_to_LVM_mapping, inplace=True)
# your code ends here
```

#### 7. Select and perform a suitable encoding method for the data (2 points)

```python
# your code starts here
import numpy as np
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

data_column_category = df.select_dtypes(exclude=[np.number]).columns
for column_name in data_column_category:
    df[column_name] = label_encoder.fit_transform(df[column_name])

df
# your code ends here
```
